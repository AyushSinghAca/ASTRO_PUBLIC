```python
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np

# Open the FITS file for getting data for our use
hdulist = fits.open("example.fits")

# optional but useful for debugging(use to see HDU)
hdulist.info()

# get our use image data from the primary HDU(Header/Data Unit) 
data = hdulist[0].data

# After geeting the data, close it.
hdulist.close()

# applying logarithmic scale as it is good for handeling negative data
safe_data = np.log(data + 1e-3)

# Display the image with a reversed 'inferno' colormap
plt.imshow(safe_data, origin="lower", cmap="inferno_r")
plt.colorbar()

# To save image
plt.savefig("fits_image.png", dpi=300, bbox_inches='tight')
plt.show()
```
